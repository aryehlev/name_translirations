```python
import numpy as np
import pandas as pd
import torch.nn as nn
import torch.optim as optim
import torch
import torch.nn.functional as F

def train_validate_test_split(dataframe, train_percent=.8,validate_percent=.1, seed=42):
    np.random.seed(seed)
    perm = np.random.permutation(dataframe.index)
   
    m = len(dataframe.index)
    train_end = int(train_percent * m)
    validate_end = int(validate_percent * m) + train_end
    print(max(perm))
    
    train = dataframe.iloc[perm[:train_end]]
    validate = dataframe.iloc[perm[train_end:validate_end]]
    test = dataframe.iloc[perm[validate_end:]]
    
    return train, validate, test
```

```python
MAX_LABEL_LENGTH = 12
MAX_INPUT_LENGTH = 14
class Vocab:
    def __init__(self, is_hebrew):
        self.char2id = {}
        self.id2char = {}
        if is_hebrew:
            self.n_chars = 2        
            self.char2id['^'] = 0
            self.char2id['$'] = 1
            self.id2char[0] = '^'
            self.id2char[1] = '$'
        else:
            self.n_chars = 0
            
        
        
        
        
    def index_text(self, text):
        indexes = [self.index_char(c) for c in text]
        return indexes
    
    def index_char(self, c):
        if c not in self.char2id:
            self.char2id[c] = self.n_chars
            self.id2char[self.n_chars] = c
            self.n_chars += 1
        return self.char2id[c]
            
            
def load_data(data, vocab_english, vocab_hebrew):
    data_sequences = []
    max_word = 0
    num_more_than_20 = 0 
    max_word_value = None
    num_english = 0
    for text in data.itertuples():
        indexes_english = vocab_english.index_text(text.english)
        indexes_hebrew = vocab_hebrew.index_text(text.hebrew)
        
        if len(indexes_hebrew) > max_word:
            max_word = len(indexes_hebrew)
            max_word_value = indexes_hebrew
        if len(indexes_hebrew) <= MAX_LABEL_LENGTH:
            for i in range(len(indexes_hebrew),MAX_LABEL_LENGTH, 1):
                indexes_hebrew.append(1)
        if len(indexes_english) <= MAX_INPUT_LENGTH:
            for i in range(len(indexes_english), MAX_INPUT_LENGTH, 1):
                indexes_english.append(1)
        if len(indexes_english) <= MAX_INPUT_LENGTH and len(indexes_hebrew) <= MAX_LABEL_LENGTH:
            data_sequences.append((indexes_english, indexes_hebrew))
        
    
    return data_sequences

vocab_english = Vocab(False)
vocab_hebrew = Vocab(True)
train_set, validation_set, test_set = train_validate_test_split(dataframe=df)

train_data = load_data(train_set,vocab_english, vocab_hebrew)

dev_data = load_data(validation_set,vocab_english, vocab_hebrew)

test_data = load_data(test_set,vocab_english, vocab_hebrew)



print(f'Train size {len(train_data)}, Dev size {len(dev_data)}, Test size {len(test_data)} vocab english size {vocab_english.n_chars} vocab hebrew size {vocab_hebrew.n_chars}')
```

```python
class GRU_encoder(nn.Module):
    def __init__(self, input_size, hidden_size, embedding_size, layers=1):
        super(GRU_encoder, self).__init__()
        self.num_of_layers = layers
        self.hidden_size = hidden_size
        self.input_size = input_size
        
        self.embedding = nn.Embedding(input_size, embedding_size).cuda()
        
        self.rnn = nn.GRU(embedding_size, hidden_size, num_layers=layers).cuda()
        
    def forward(self, input_letter, hidden):
        output = self.embedding(input_letter)
        
        
        
        output, hidden = self.rnn(output, hidden)
        
        return output, hidden
        
    def initHidden(self):
        return torch.zeros(self.num_of_layers, 1, self.hidden_size).cuda()  
```

```python
class GRU_decoder(nn.Module):
    def __init__(self,  output_size, hidden_size, embedding_size, layers=1, dropout=0.1):
        super(GRU_decoder, self).__init__()
        self.hidden_size = hidden_size
        self.output_size = output_size
        self.dropout = dropout
        self.num_of_layers = layers
        

        self.embedding = nn.Embedding(self.output_size, self.hidden_size).cuda()
        self.linear = nn.Linear(self.hidden_size * 2, MAX_INPUT_LENGTH).cuda()
        
        self.linear_layers = nn.Linear(self.num_of_layers * MAX_INPUT_LENGTH, MAX_INPUT_LENGTH).cuda()
        
        self.linear_attention = nn.Linear(self.hidden_size * 2, self.hidden_size).cuda()
        self.dropout = nn.Dropout(self.dropout).cuda()
        self.gru = nn.GRU(self.hidden_size, self.hidden_size, num_layers=layers).cuda()
        self.out = nn.Linear(self.hidden_size, self.output_size).cuda()

    def forward(self, letter, hidden, encoder_outputs):
        embedded = self.embedding(letter)
#         embedded = self.dropout(embedded)
        
#         output = self.embedding(letter)
#         output = self.dropout(output)
        
#         print(embedded.size())
#         print(hidden.size())
        sum_of_layers = torch.zeros(1, self.hidden_size)
        
        weights_layers = torch.empty(self.num_of_layers * MAX_INPUT_LENGTH).cuda()
        
        for i in range(self.num_of_layers):
            
            output = torch.cat((embedded[0][0], hidden[i][0]))
            output = self.linear(output)
    #             print(output.size())
            weights = F.softmax(output) 
            
#             print(weights.size())
    #             print(output.size())
            if i == 0:
                weights_layers = weights
            else:
                weights_layers = torch.cat((weights_layers, weights))

    #             weights = F.softmax(
    #                 self.attn(torch.cat((output[0], hidden[0]), 1)), dim=1)
    #             print(encoder_outputs.size())
#         print(weights_layers.size())
#         print(self.num_of_layers * MAX_INPUT_LENGTH)
        weights = self.linear_layers(weights_layers)
        encoder_outputs = encoder_outputs.reshape(encoder_outputs.shape[0], encoder_outputs.shape[2])
#             print(len(weights))
        
        
        weights = weights.reshape(1, len(weights))
        attn_applied = torch.mm(weights, encoder_outputs)

#             print(attn_applied.size())
#             print(embedded.size())
#             print(attn_applied.size())
#             print(attn_applied.size())
        output = torch.cat((embedded[0][0], attn_applied[0]))
#         print(output.size())
        output = self.linear_attention(output)

        output = F.relu(output)
        output = output.reshape(1,1,self.hidden_size)
        
        output, hidden = self.gru(output, hidden)
        
        output = self.out(output)
#         print(output.size())
        return output, hidden
```

```python
encoder_model_3 = GRU_encoder(vocab_english.n_chars, 300, 64, layers=2)
decoder_model_3 = GRU_decoder(vocab_hebrew.n_chars, 300, 64, layers=2)

encoder_optimizer_3 = optim.Adam(params=encoder_model_3.parameters(), lr=0.0001)
decoder_optimizer_3 = optim.Adam(params=decoder_model_3.parameters(), lr=0.0001)

critertion_3 = nn.CrossEntropyLoss()
```

```python


def train_3(n_epochs=4, print_every=5000):
    loss = 0

    for e in range(1, n_epochs + 1):

        for counter, (english_name, hebrew_name) in enumerate(small_train_data):
            encoder_hidden = encoder_model_3.initHidden()

            encoder_optimizer_3.zero_grad()
            decoder_optimizer_3.zero_grad()

    #         english_name_tensor = torch.LongTensor(english_name).cuda()
            encoder_outputs = torch.ones(MAX_INPUT_LENGTH, 1, encoder_model_3.hidden_size).cuda()
    #         hebrew_name_tensor = torch.LongTensor(hebrew_name).cuda()
    
            for i, letter in enumerate(english_name):
                letter_tensor = torch.tensor([[letter]]).cuda()
                encoder_output, encoder_hidden = encoder_model_3(letter_tensor, encoder_hidden)
                
                encoder_outputs[i] = encoder_output
            
            decoder_input = torch.tensor([[0]]).cuda()
            decoder_hidden = encoder_hidden
            
            transliration_loss = 0 
            
            
            for i in range(len(hebrew_name)):
                decoder_output, decoder_hidden = decoder_model_3(decoder_input, decoder_hidden, encoder_outputs)
#                 print(i)
                
                
                _, top_indexes = decoder_output.topk(1)

                decoder_input = top_indexes.squeeze(0).detach()
#                 return
                
#                 print(decoder_hidden.size())
                transliration_loss += critertion_3(decoder_output.reshape(decoder_output.shape[1], decoder_output.shape[2]), torch.tensor([hebrew_name[i]]).cuda())
                if decoder_input.item() == [1]:
                    break
#             return 
            transliration_loss.backward()

            encoder_optimizer_3.step()
            decoder_optimizer_3.step()

            loss += transliration_loss.item() / len(hebrew_name)


            if counter % print_every == 0 and counter != 0:
                loss = loss / print_every
                print('Epoch %d,/%d, Current Loss = %.4f' % (e, counter,loss))
                loss = 0

```
